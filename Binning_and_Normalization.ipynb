```python
import pandas as pd
import numpy as np

df = pd.read_excel('/content/auto-mpg .xlsx')
df.head()
df.drop(['origin'], axis = 1)
```

```python
df.isnull().sum()
```

```python
avg_acc = df['acceleration'].mean()
print(avg_acc)
```

```python
df['acceleration'].replace( np.nan, avg_acc , inplace = True)
```

```python
print(df[df['model year'].isnull()])
```

```python
df.isnull().sum()
```

```python
model_year_mc=df['model year'].value_counts().idxmax()
```

```python
df['model year'].replace( np.nan, model_year_mc , inplace = True)
```
